```python
from pymongo import MongoClient

# Connexion au serveur MongoDB local
client = MongoClient("mongodb://localhost:27017/")

# Sélection de la base de données
db = client["nocites"]

# Sélection de la collection Paris
collection = db["listings_paris"]
```

```python
import polars as pl

# Charger toutes les données de la collection MongoDB dans une liste
data = list(collection.find())

# Nettoyer les documents pour remplacer les chaînes vides par None
# Pourquoi ? 
# Polars a besoin que chaque colonne ait un type de données cohérent. 
# Si certaines lignes contiennent des chaînes vides "" et d'autres des nombres ou des dates, 
# Polars ne peut pas déterminer un type unique et lève une erreur 
# comme "could not append value: '' of type: str to the builder".
# En remplaçant "" par None (valeur manquante), Polars peut correctement inférer le type 
# et créer le DataFrame sans erreurs.
def clean_doc(doc):
    return {k: (v if v != "" else None) for k, v in doc.items()}

data_clean = [clean_doc(d) for d in data]

# Créer un DataFrame Polars
df = pl.from_dicts(data_clean)

# Vérifier les premières lignes
df.head()

```

```python
# -------------------------------
# 1. Calculer le taux de réservation moyen sur 30 jours par type de logement
# -------------------------------
# Objectif : savoir combien de jours en moyenne les logements sont réservés, par type
# Taux de réservation = ((30 - disponibilité sur 30 jours) / 30) * 100

# Créer une colonne 'reservation_rate_30' qui calcule le taux de réservation
df = df.with_columns([
    ((30 - pl.col("availability_30").fill_null(0)) / 30 * 100).alias("reservation_rate_30")
])

# Grouper par 'room_type' et calculer la moyenne du taux de réservation
avg_reservation = df.group_by("room_type").agg([
    pl.mean("reservation_rate_30").alias("avg_reservation_rate")
])

# Afficher le résultat
avg_reservation
```

```python
# -------------------------------
# 2. Calculer la médiane du nombre d'avis pour tous les logements
# -------------------------------
# Objectif : obtenir la valeur centrale du nombre d'avis
# Pourquoi la médiane : elle n'est pas influencée par les extrêmes (quelques logements avec très beaucoup d'avis)

median_reviews = df.select([
    pl.col("number_of_reviews").median().alias("median_reviews")
])

# Afficher le résultat
median_reviews
```

```python
# -------------------------------
# 3. Calculer la médiane du nombre d'avis par catégorie d'hôte (superhost ou non)
# -------------------------------
# Objectif : comparer les superhosts aux autres hôtes
# host_is_superhost = 't' pour superhost, 'f' pour les autres

median_reviews_host = df.group_by("host_is_superhost").agg([
    pl.col("number_of_reviews").median().alias("median_reviews")
])

median_reviews_host
```

```python
# -------------------------------
# 4. Calculer la densité de logements par quartier
# -------------------------------
# Objectif : savoir combien de logements sont présents par quartier
# Cela aide à identifier les quartiers les plus denses en locations

density_per_quartier = df.group_by("neighbourhood_cleansed").agg([
    pl.len().alias("nb_logements")
]).sort("nb_logements", descending=True)  # Tri pour afficher les quartiers les plus denses en premier

density_per_quartier
```

```python
# -------------------------------
# 5. Identifier les quartiers avec le plus fort taux de réservation
# -------------------------------
# Objectif : identifier les quartiers où les logements sont le plus souvent réservés
# Taux de réservation déjà calculé dans 'reservation_rate_30'

top_quartiers = df.group_by("neighbourhood_cleansed").agg([
    pl.mean("reservation_rate_30").alias("avg_reservation_rate")
]).sort("avg_reservation_rate", descending=True)

# Afficher les 5 quartiers avec le plus fort taux de réservation
top_quartiers.head(5)
```
